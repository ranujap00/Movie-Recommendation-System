```python
import requests
import os

# TMDb API endpoint for movie search
api_url = "https://api.themoviedb.org/3/search/movie"

# TMDb API key (you need to get your own key from TMDb)
api_key = "596cc532fc70025e472ef8c89d5ae4a7"
```

```python
import pandas as pd

movies = pd.read_csv('movies.csv')
movie_titles = list(movies['title'])
print(len(movie_titles))

# movie_titles = movie_titles[1:20]
movie_titles
```

```python
movie_data = []

for movie_title in movie_titles:
    # Remove the year from the movie title
    movie_title = movie_title.split(' (')[0]

    # Make a request to TMDb API to search for the movie
    params = {
        "api_key": api_key,
        "query": movie_title
    }
    response = requests.get(api_url, params=params)
    data = response.json()
    
    if data["results"]:
        # Get the poster path for the first result
        poster_path = data["results"][0]["poster_path"]
        
        # Construct the URL for the poster and add it to the list
        poster_url = f"https://image.tmdb.org/t/p/original{poster_path}"
        movie_data.append({"Title": movie_title, "Poster URL": poster_url})
    else:
        # If poster not found, add only the title to the list
        movie_data.append({"Title": movie_title, "Poster URL": "Not found"})

```

```python
movie_data_df = pd.DataFrame(movie_data)
movie_data_df = movie_data_df.rename(columns={'Title': 'title'})
movie_data_df
```

```python
# Create a new Series with the movie titles
movie_titles_series = pd.Series(movie_titles, name='title')

# Replace the "Title" column in the DataFrame with the new Series
movie_data_df['title'] = movie_titles_series
movie_data_df
```

```python
# merge dfs
merged_df_url = movies.merge(movie_data_df, how='inner', on='title')
merged_df_url
```

```python
## write to a csv file
merged_df_url.to_csv('movie_data_with_urls.csv', index=False)  # Set index=False to exclude the index column
```
